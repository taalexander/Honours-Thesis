```python
models = np.linspace(0,1,100)
prior = np.empty(100)
prior.fill(0.01)
events = np.empty(10)
events.fill(1)
```

```python
def prob(bias,events):
    pr = 1 
    for x in events: 
        if x==0:
            pr*=bias  
        else:
            pr*= (1-bias)
    return pr 
def bayes (models,priors,events):
    posteriors = np.empty(len(models))
    posteriors.fill(0)
    normal = 0
    for i,x in enumerate(models): 
        normal += prob(x,events)*priors[i]
    for i,x in enumerate(models): 
        posteriors[i] = prob(x,events)*priors[i]/normal 
    return posteriors
```

```python
after = bayes(models,prior,events)
```

```python
plot(models,prior)
```

```python
plot(models,after)
```

```python
fig = figure(1)
plt.subplot(211)
vlines(models,0,prior,"b",lw=2)
ylim(0,0.12)
xlim(-0.01,1.01)
title("Prior Model Distribution") 

ylabel("Model Liklihood")
plt.subplot(212)
fig.tight_layout()
vlines(models,0,after,"b",lw=2)
ylim(0,0.12)
xlim(-0.01,1.01)
title("Posterior Model Distribution After 20 Consecutive Heads") 
xlabel("Coin Bias (zero all heads)")
ylabel("Model Liklihood")
fig.savefig("bayes.eps")
plt.show()
```
